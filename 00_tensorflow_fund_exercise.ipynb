##  7. Find the min and max values of the tensor you created in 6 along the first axis.

```python
matrix
```

```python
import numpy as np
import tensorflow as tf

numpy_a = np.arange(1, 37, dtype=np.int32)  # 36 elements
tensor_a = tf.constant(numpy_a, shape=[2, 6, 3])
print(tensor_a)
```

```python
# Find minimum value
min_value = tf.reduce_min(tensor_a)

# Find maximum value
max_value = tf.reduce_max(tensor_a)

print("Min:", min_value.numpy())
print("Max:", max_value.numpy())
```

```python
import numpy as np
import tensorflow as tf
tf.random.set_seed(42)
tensor_10a=tf.constant(np.arange(2,62, dtype=np.int16), shape=(2,3,10))
tensor_10a
```

```python
tf.reduce_min(tensor_10a).numpy(), tf.reduce_max(tensor_10a).numpy()
```

```python
# index of the maximum value
tf.where(tensor_10a == tf.reduce_max(tensor_10a))
```

```python
# index of the minimum value
tf.where(tensor_10a == tf.reduce_min(tensor_10a))
```
